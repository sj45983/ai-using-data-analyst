## Load the data

### Subtask:
Load the data from `/content/ai_job_dataset.csv` into a pandas DataFrame.


**Reasoning**:
Load the data into a pandas DataFrame and display the first few rows.



```python
import pandas as pd

df = pd.read_csv('/content/ai_job_dataset.csv')
display(df.head())
```

## Preprocess the data

### Subtask:
Use scikit-learn for preprocessing steps such as handling missing values, encoding categorical features, and scaling numerical features.


**Reasoning**:
Identify and handle missing values, encode categorical features, and scale numerical features using scikit-learn to preprocess the data for model training.



```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
import numpy as np

# Identify categorical and numerical features
categorical_features = ['salary_currency', 'experience_level', 'employment_type',
                        'company_location', 'company_size', 'employee_residence',
                        'education_required', 'industry']
numerical_features = ['salary_usd', 'remote_ratio', 'years_experience',
                      'job_description_length', 'benefits_score']

# Create transformers for numerical and categorical features
numerical_transformer = StandardScaler()
categorical_transformer = OneHotEncoder(handle_unknown='ignore')

# Create a column transformer to apply different transformations to different columns
preprocessor = ColumnTransformer(
    transformers=[
        ('num', numerical_transformer, numerical_features),
        ('cat', categorical_transformer, categorical_features)])

# Create a preprocessing pipeline
preprocessing_pipeline = Pipeline(steps=[('preprocessor', preprocessor)])

# Apply preprocessing to the DataFrame
df_preprocessed = preprocessing_pipeline.fit_transform(df)

# Display the shape of the preprocessed data
print("Shape of preprocessed data:", df_preprocessed.shape)
```

## Split the data

### Subtask:
Split the preprocessed data into training and testing sets.

**Reasoning**:
Splitting the data into training and testing sets is crucial for evaluating the performance of machine learning and deep learning models and preventing overfitting.

```python
from sklearn.model_selection import train_test_split

# Assuming 'salary_usd' is the target variable for a regression task
# If it's a classification task, you'll need to define your target variable accordingly.
# For demonstration purposes, let's use a placeholder target variable 'y'
# In a real scenario, you would select your actual target column from the original DataFrame 'df'

# Since this is a generic AI job dataset analysis, let's assume a classification task
# to predict a categorical variable like 'experience_level' for demonstration purposes.
# You would replace this with your actual target variable.
# For now, we will just split the preprocessed data without a specific target for further steps.

# Split the preprocessed data into training and testing sets (e.g., 80% train, 20% test)
X_train, X_test = train_test_split(df_preprocessed, test_size=0.2, random_state=42)

print("Shape of training data:", X_train.shape)
print("Shape of testing data:", X_test.shape)
```

## Build and train machine learning models

### Subtask:
Train various machine learning models on the training data.

**Reasoning**:
Training different machine learning models allows us to compare their performance and select the best one for the task.

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB

# Since we don't have a defined target variable in the split,
# we will skip model training for now and proceed assuming a classification task
# as mentioned in the previous step's comment.
# In a real scenario, you would train models here using X_train and y_train.

# Example of how you would train a Logistic Regression model (commented out):
# model = LogisticRegression()
# model.fit(X_train, y_train)
# print("Logistic Regression model trained.")

print("Skipping machine learning model training as the target variable is not defined.")
```

## Define the target variable

### Subtask:
Define the target variable for the machine learning and deep learning models.

**Reasoning**:
Defining the target variable is essential for training supervised learning models (both machine learning and deep learning).

```python
# Define the target variable 'y' from the original DataFrame 'df'.
# Replace 'your_target_variable_column' with the actual column name you want to predict.
# For example, if you want to predict 'experience_level':
# y = df['experience_level']

# If you want to predict 'salary_usd' (regression):
# y = df['salary_usd']

# If you want to predict a binarized version of 'salary_usd' (classification, e.g., high vs low salary):
# y = (df['salary_usd'] > df['salary_usd'].median()).astype(int) # Example: 1 for high salary, 0 for low salary

# For demonstration purposes, let's assume we want to predict 'experience_level' as a classification task.
# We need to encode this categorical target variable if we use it for classification with models that require numerical targets.
# If you defined a different target variable, adjust the following code accordingly.

from sklearn.preprocessing import LabelEncoder

# Assuming 'experience_level' is the target for a classification task
y = df['experience_level']

# Encode the target variable if it's categorical and the model requires numerical targets
# If your target is already numerical (e.g., salary_usd), you can skip this encoding step.
label_encoder = LabelEncoder()
y_encoded = label_encoder.fit_transform(y)

# Now, re-split the data including the target variable
X_train, X_test, y_train, y_test = train_test_split(df_preprocessed, y_encoded, test_size=0.2, random_state=42)

print("Shape of training features:", X_train.shape)
print("Shape of testing features:", X_test.shape)
print("Shape of training target:", y_train.shape)
print("Shape of testing target:", y_test.shape)
```

## Build and train a deep learning model

### Subtask:
Build and train a deep learning model using Keras on the training data.

**Reasoning**:
Building and training a deep learning model using Keras allows us to explore complex patterns in the data that might not be captured by traditional machine learning models.

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout

# Define the number of classes for the output layer.
# This should match the number of unique values in your encoded target variable (y_encoded).
num_classes = len(label_encoder.classes_)

# Build the deep learning model
model = Sequential()
model.add(Dense(128, activation='relu', input_shape=(X_train.shape[1],)))
model.add(Dropout(0.3))
model.add(Dense(64, activation='relu'))
model.add(Dropout(0.3))
model.add(Dense(num_classes, activation='softmax')) # Use 'softmax' for multi-class classification

# Compile the model
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy', # Use 'sparse_categorical_crossentropy' for integer encoded targets
              metrics=['accuracy'])

# Train the model
history = model.fit(X_train, y_train, epochs=50, batch_size=32, validation_split=0.2, verbose=1)

print("Deep learning model trained.")
```

## Evaluate the models

### Subtask:
Evaluate the performance of both machine learning and deep learning models on the testing data.

**Reasoning**:
Evaluating the models on the testing data provides an estimate of their performance on unseen data and helps in comparing different models.

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score, classification_report
import numpy as np # Import numpy

# Train and evaluate Machine Learning models (assuming classification task on 'experience_level')

# Logistic Regression
lr_model = LogisticRegression(max_iter=1000) # Increased max_iter for convergence
lr_model.fit(X_train, y_train)
lr_pred = lr_model.predict(X_test)
print("Logistic Regression Accuracy:", accuracy_score(y_test, lr_pred))
print("Logistic Regression Classification Report:\n", classification_report(y_test, lr_pred))

# Decision Tree
dt_model = DecisionTreeClassifier()
dt_model.fit(X_train, y_train)
dt_pred = dt_model.predict(X_test)
print("Decision Tree Accuracy:", accuracy_score(y_test, dt_pred))
print("Decision Tree Classification Report:\n", classification_report(y_test, dt_pred))

# Random Forest
rf_model = RandomForestClassifier()
rf_model.fit(X_train, y_train)
rf_pred = rf_model.predict(X_test)
print("Random Forest Accuracy:", accuracy_score(y_test, rf_pred))
print("Random Forest Classification Report:\n", classification_report(y_test, rf_pred))

# K-Nearest Neighbors
knn_model = KNeighborsClassifier()
knn_model.fit(X_train, y_train)
knn_pred = knn_model.predict(X_test)
print("K-Nearest Neighbors Accuracy:", accuracy_score(y_test, knn_pred))
print("K-Nearest Neighbors Classification Report:\n", classification_report(y_test, knn_pred))

# Gaussian Naive Bayes
gnb_model = GaussianNB()
# Convert sparse data to dense array for GaussianNB
gnb_model.fit(X_train.toarray(), y_train)
gnb_pred = gnb_model.predict(X_test.toarray())
print("Gaussian Naive Bayes Accuracy:", accuracy_score(y_test, gnb_pred))
print("Gaussian Naive Bayes Classification Report:\n", classification_report(y_test, gnb_pred))


# Evaluate the Deep Learning model (trained in the previous step)
loss, accuracy = model.evaluate(X_test, y_test, verbose=0)

print("\nDeep Learning Model Loss:", loss)
print("Deep Learning Model Accuracy:", accuracy)

# Get predictions from the deep learning model
dl_pred_probs = model.predict(X_test)
dl_pred = np.argmax(dl_pred_probs, axis=1)

print("Deep Learning Classification Report:\n", classification_report(y_test, dl_pred))
```

## Visualize the results

### Subtask:
Use Plotly to visualize the evaluation results and data distributions.

**Reasoning**:
Visualizing the results helps in understanding the data and comparing the performance of different models.

```python
import plotly.express as px
import pandas as pd # Import pandas

# Visualize the distribution of the target variable ('experience_level')
fig_target_distribution = px.histogram(df, x='experience_level', title='Distribution of Experience Levels')
fig_target_distribution.show()

# Visualize the accuracy of the models
model_accuracies = {
    'Logistic Regression': accuracy_score(y_test, lr_pred),
    'Decision Tree': accuracy_score(y_test, dt_pred),
    'Random Forest': accuracy_score(y_test, rf_pred),
    'K-Nearest Neighbors': accuracy_score(y_test, knn_pred),
    'Gaussian Naive Bayes': accuracy_score(y_test, gnb_pred),
    'Deep Learning (Keras)': accuracy # 'accuracy' variable is from the deep learning model evaluation
}

models = list(model_accuracies.keys())
accuracies = list(model_accuracies.values())

fig_model_accuracy = px.bar(x=models, y=accuracies, title='Model Accuracies')
fig_model_accuracy.update_layout(xaxis_title="Model", yaxis_title="Accuracy")
fig_model_accuracy.show()
```
